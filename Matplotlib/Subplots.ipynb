## Visualisation in Python - Matplotlib

### Subplots: Comparing the trend of sales across different regions

-  Plotting different categories in the same graph or next to each other
-  Helps in comparing and summarising the information over multiple categories

```python
# importing numpy and the pyplot package of matplotlib
import numpy as np
import matplotlib.pyplot as plt
# Data for sales made by the company across three regions over the years 2012 to 2015
years = np.array(['2012', '2013', '2014', '2015'])
sales_Africa = np.array([127187.27, 144480.70, 229068.79, 283036.44])
sales_Asia_Pacific = np.array([713658.22, 863983.97, 1092231.65, 1372784.40])
sales_Europe = np.array([540750.63, 717611.40, 848670.24, 1180303.95])
```

```python
# Plotting the trend of sales in all three regions over the 4 years using subplots

fig,ax=plt.subplots()#subplots return two attributes

#we put the comma becoz plot method returns two attributes
africa,=ax.plot(years,sales_Africa)
africa.set_label("Africa")
africa.set_dashes([2,2])

asia,=ax.plot(years,sales_Asia_Pacific)
asia.set_label("Asia Pacific")

europe,=ax.plot(years,sales_Europe)
europe.set_label("Europe")
#We wants to represent the europe data in different form
europe.set_dashes([2,2,5,2])
#its mean that ([dot upto two points,gap of two points,dash upto 5 points,gap upto 2 points ... and repeat same])

plt.title("Comparing the trend of sales across different regions")
plt.xlabel("Years")
plt.ylabel("Sales")

#lets make sales (y-axis as more attractive way) like 2,00,000 - 2L(L=lakhs)
tick_sales=np.arange(0,1600000,200000)
tick_labels=list()
for i in range(8):
    dis=i*2
    text=str(dis)+'L'
    tick_labels.append(text)
plt.yticks(tick_sales,tick_labels)

plt.legend()

plt.show()
```

#### Subplots with a shared y-axis

```python
#Here we need to tell in how many parts it will divided ncol=3
#sharey means we are sharing the y axis labels with all the divided graph to look in better way
fig,ax=plt.subplots(ncols=3,sharey=True)

#the subgraphs are stored as the array in the ax so we need to call it by the help of the index (0-2)
africa,=ax[0].plot(years,sales_Africa)
africa.set_label("Africa")
africa.set_dashes([2,2])

asia,=ax[1].plot(years,sales_Asia_Pacific)
asia.set_label("Asia Pacific")

europe,=ax[2].plot(years,sales_Europe)
europe.set_label("Europe")
#We wants to represent the europe data in different form
europe.set_dashes([2,2,5,2])
#its mean that ([dot upto two points,gap of two points,dash upto 5 points,gap upto 2 points ... and repeat same])

plt.title("Comparing the trend of sales across different regions")
plt.xlabel("Years")
plt.ylabel("Sales")

#lets make sales (y-axis as more attractive way) like 2,00,000 - 2L(L=lakhs)
tick_sales=np.arange(0,1600000,200000)
tick_labels=list()
for i in range(8):
    dis=i*2
    text=str(dis)+'L'
    tick_labels.append(text)
plt.yticks(tick_sales,tick_labels)

#But here we need to change the syntex of the legend() method
plt.legend([africa,asia,europe],["Africa","Asia Pacific","Europe"])

plt.show()
```
