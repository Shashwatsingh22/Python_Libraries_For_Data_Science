```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
product_cat=np.array(["Furniture","Technology","Office Supplies"])
sales=np.array([4110454.5,5842562.2,4110454.49])
```

```python
plt.bar(product_cat,sales,color="orange",edgecolor="Black")
plt.title("Graph Between the Sales And The Products Categories",color="Blue")
plt.xlabel("Product Category",color="Green")
plt.ylabel("Sales",color="Green")

y_axis=np.arange(0,7000000,1000000)
y_txt=[]
for i in range(8):
    txt=str(i)+"L"
    y_txt.append(txt)
plt.yticks(y_axis,y_txt)

plt.show()
```
