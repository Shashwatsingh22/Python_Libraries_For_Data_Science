## Printing The Rows Based On the Condition
### We want to Select row which satisfy some given conditions

```python
df[["Profit"]]<0
```

```python
#It is priting in the form of the Bool ... lets print in the form of the value
df[df["Profit"]<0]
```

```python
# Lets Apply the Multiple Conditions
df[(df["Sales"]>100000) & (df["Market"]=="LATAM") | (df["Market"]=="Europe")]
```

```python
# Lets Improve the Syntax in better format...
#---> We can use ("isin()") instead of the use of ("|") or symbol...
# When we use isin() method we have to give some set of the Strings then isin() method will use to see the whole list of
# the string if any string matches it will throw our output....
df[(df["Sales"]>100000) & (df["Market"].isin(["LATAM","Europe"]))]
```
