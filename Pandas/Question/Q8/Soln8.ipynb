## Dataframe Pivot Table
### Description
#### Group the data 'df' by 'month' and 'day' and find the mean value for column 'rain' and 'wind' using the pivot table command.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("forestfires.csv",index_col=[2,3])
```

```python
df_1=pd.pivot_table(data=df, index=["month","day"], values=["rain","wind"], aggfunc='mean')
```
