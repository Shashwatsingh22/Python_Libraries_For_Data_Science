## Dataframes Merge
#### Description
##### Perform an inner merge on two data frames df_1 and df_2 on  'unique_id' and print the combined dataframe.

```python
import numpy as np 
import pandas as pd
```

```python
df_1=pd.read_csv("restaurant-1.csv")
df_2=pd.read_csv("restaurant-2.csv")
df_1.head(3)
```

```python
df_2.head(3)
```

```python
df_3=df_1.merge(df_2,on="unique_id",how="inner")
df_3.head(20)
```
