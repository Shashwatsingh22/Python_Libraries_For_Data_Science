## Dataframe grouping
### Description
#### Group the dataframe 'df' by 'month' and 'day' and find the mean value for column 'rain' and 'wind'.

```python
import numpy as np
import pandas as pd
```

```python
df =pd.read_csv("forestfires.csv",index_col=[2,3])
df
```

```python
df_2=df[["rain","wind"]].groupby(["month","day"]).mean()
df_2.head(20)
```
