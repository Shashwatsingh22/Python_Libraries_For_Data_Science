# Handling Time - Series Data
### Time series datas are reffers to as the data which are store with or according to the time.Data  are stored at regular time interval........Examples->Weather Report, Stock Prices , Temperature etc.... as intormation look logical when these reports are only present with or according to time interval....

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Information About Datasets:
### Weather Report--- Austin City    2010: 1st Jan to 31st Dec    Attributes--Temp,Dew Point,Pressure,Date And Time      Data is recorded at every Hour

```python
df=pd.read_csv("DataSets\weather_data.csv")
df
```

```python
#As we can see in the col of the Date_time there values or we can say the date and time formate are not in suitable manner
#they all are in string format, so lets convert it in Date_Fromate
df=pd.read_csv("DataSets/weather_data.csv",parse_dates=["Date_Time"])
df
```

```python
#Now lets make the Date_Time as the row header and also lets give the value to the Parse_Date as True
df=pd.read_csv("DataSets/weather_data.csv",parse_dates=True,index_col="Date_Time")
df
```

```python
#Lets see How we can slice the Data from the DataFrame according to the Time...
#We wants to fetch the data for 5th feb 2010
df.loc["February 5, 2010"]
```

```python
#Fetch data=> 9 Am to 11 Am on 11 Oct 2010
df.loc["2010-10-11 9:00:00":"2010-10-11 11:00:00"]
```

```python
#Fetch Data=> 15 dec to 31 dec
plot_line=df.loc["2010-12-15":"2010-12-31"]
plot_line
```

```python
#Lets Plot the Data--> So, as we know that line plot is the best plot to analyse data over time.
#Ploting the data from 15 Dec to 31 Dec v/s Temp.
# First will fetch the temp. data..
plot_line["Temperature"].plot(kind="line")
plt.show()
```
