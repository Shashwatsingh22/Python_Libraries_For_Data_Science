## Merge & Append
### In the real world secnerio , we would find rarely have the complete data stored in single table to load into dataframe.We have to collect the table from different - different source and then by the help python we can Merge or Append or Concatenate them to form a complete DataFrame...

```python
#Lets take an example of such type of secnerio like we have data which holds the customer accept the order and 
# also have data in which customer return the order...
import numpy as np
import pandas as pd
```

```python
#So, as we know that we have on file of ext-> .xlsx it have both the data of Oreders as well as Retuens ,
#then How we can tell this thing pandas ? -> It will done by an attribut is "sheet_name=xyz"
accept=pd.read_excel("DataSets/sales_returns.xlsx",sheet_name="Orders")
returns=pd.read_excel("DataSets/sales_returns.xlsx",sheet_name="Returns")
```

```python
print(accept.shape)
accept.head()
```

```python
print(returns.shape)
returns.head()
```

```python
#Lets merge these two dataframs ...for merging we need to observe that there should be one common factor ...
#So, as per the given dataframe we have orderID which common in both of them....and we need to specify also by the help of
#an attribute "on=" which will be pass under the method of "merge()"
#Here we are creating a new dataframe...
merge_df=accept.merge(returns,on="Order ID")
print(merge_df.shape)
merge_df.head(20)
```

```python
# the no. of the rows are same as the datfram which we have "return"...and this type of join is known as "inner join"
#But we won't like this we want that those who accepted it shows --> No, and those Not--->Yes...so we need to take an help
# of the attribute called "how="...
merge_df=accept.merge(returns,on="Order ID",how="left")
merge_df.head(60)
```

```python
#As you can see that I have specified as the left which mean "Left Join"
#There 4 types of merging is there:-
#1>.left-->this will select entries only in the 1st DataFrame.
#2>.right-->select entries only in the 2nd DataFrane.
#3>.outer-->union of all the entries
#4>.inner-->intersection of the keys from both  the dataframe
#Instead of NaN we wants Yes or No...
#First we need to change the col of the Returned in string...
merge_df["Returned"]=merge_df["Returned"].astype("str")

#lets apply the Lambda function..
merge_df["Returned"]=merge_df["Returned"].apply(lambda x:"No" if x=="nan" else "Yes")
merge_df.head(60)
```

## Concatenating DataFrames:
### Concatenation is used when we have dataframes with the same columns and wants to merge them on the top of eachothe or with the same rows and wants to append them side by side...

```python
#Lets take an example...
df_1=pd.DataFrame({"Name":["Sain","Aman","Raj","Saif"],
                  "Age":["34","31","22","33"],
                  "Gender":["M","M","M","M"]})
df_2=pd.DataFrame({"Name":["Asha","Preeti","Ansh"],
                  "Age":["31","22","23"],
                  "Gender":["F","F","M"]})
#As we can see that in these two dataframes the columns are same so we need to merge them on the top of eachother...
df_1
```

```python
df_2
```

```python
#lets append them
df_1.append(df_2)
```

```python
df_3=pd.DataFrame({"School":["DAV","DPS","UPS","JSP"],
      "Marks":["84","89","76","91"]})
#As we can see that it have two different columns , so we need to merge it as the seprate col.
df_3
```

```python
pd.concat([df_1,df_3])
```

```python
#The o/p is giving some NaN data.. So, the concat function is appending(On top of eachother) by default we wan them as
#Concatenate them from side by side we need to take an help of an attribute
#"axis=0(ByDefult it is 0 which means appending on top of eachother)"
#"axis=1(It means side by side)"
pd.concat([df_1,df_3],axis=1)
```
