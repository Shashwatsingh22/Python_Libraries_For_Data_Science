```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel("DataSets/Sales.xlsx")
df
```

```python
#As we have loaded the Data is the values which are provide is it neccsary that Data is valid to Use ?
#NO, We need to do some changes in it.
#Lets see our example of the dataset as we can see that the sales value its huge value in Lakhs, It will becomes very 
#dificult to read those value....
#Now , Lets try to reduce those values into the multiples of the Thousands
#Like 656637/1000 = 656.637 K (Lets Change the Value Inform of the Thousands)
df.Sales=df.Sales.floordiv(1000)
#df.Sales it is another way to extract a single col...

df
```

```python
#We have Modifies the Value But the Heading remains same (Its shows that we have lost our previous Data)
#Lets Solve this problem by rename the header
df.rename(columns={"Sales":"Sales in Thousands"}).head()
#col will take it as dict and key will be here is new name of header of that col
```

```python
#Now it is returning us the new data frame
#Lets Proof it by printing the original one....
df.head()
```

```python
#It is an problem to us... There is an attribute in method/function in rename() "inplace=False" byDefault
#Bascially this attribute determines that renameing will be applied on the current datafram or rename to new dataframe
#False--->New DataFrame
#True--->Current DataFrame
df.rename(columns={"Sales":"Sales In Thousands"},inplace=True)
df.head()
```

## Lambda Function
#### It is an another way to modify or create a new column by using the Lambda Function....                                                                                        Lets in our example of the datasets we wants to create a seprate col for the Positive_Profit....So, the column which created is known as "Derived Variable".

```python
#It will be done by the apply() method on column of the DataFrame...We give in place of negative is NaN
df["+ve_Profit"]=df.Profit.apply(lambda x : np.nan if x<0 else x)
df.head()
```
